```r
rm(list=ls())
library(data.table)
library(ggplot2)
library(repr)
```

```r
log=nodup_log=read.table("6kb/k562.h3k27ac.seed.1234.cs.25.filters.500.naive.range.4.6.to.11.5.0.log.batch",sep='\t',header=TRUE)
weight=25
log$Batch=seq(1,nrow(log))
log$scaledcounts=weight*log$logcount_predictions
options(repr.plot.width=20, repr.plot.height=5)
p1=ggplot(data=log)+
geom_line(aes(x=Batch,y=loss,color="Training_Total"))+
geom_line(aes(x=Batch,y=logcount_predictions_loss,color="Training_LogCount"))+
geom_line(aes(x=Batch,y=scaledcounts,color="Training_LogCount_x25"))+
geom_line(aes(x=Batch,y=profile_predictions_loss,color="Training_Profile"))+
xlab("Batch")+
ylab("Loss")+
ggtitle("K562 H3K27ac\nBPNet w/500 Filters\nWith duplicates, count loss x 25")+
scale_color_manual(name="Loss",
      values = c(
           Training_LogCount='#1b9e77',
           Training_LogCount_x25='#000000',
           Training_Profile='#d95f02',
           Training_Total='#7570b3'))+  
scale_y_continuous(trans = 'log2')+
theme_bw(20)
p1
```
