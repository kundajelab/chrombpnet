```python
import pandas as pd
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("c",sep=",")
```

```python
val = data["chrombpnet model performance peaks (mse counts)"]
means = [float(y.split(" ")[0]) for y in val]
stds = [float(y.split(" ")[1][1:-1]) for y in val]
```

```python
plt.plot(data["dilation layer"], means,c="r",alpha=0.5)
plt.errorbar(x=data["dilation layer"], y=means, yerr=stds, fmt = 'o',color = 'red', 
            ecolor = '#ffcccb', elinewidth = 5, capsize=10)
plt.ylabel("Mean squared error")
plt.xlabel("Dilation Layers")
```

```python
val = data["chrombpnet model performance peaks (pearsonr counts)"]
means = [float(y.split(" ")[0]) for y in val]
stds = [float(y.split(" ")[1][1:-1]) for y in val]
```

```python
plt.plot(data["dilation layer"], means,c="r",alpha=0.5)
plt.errorbar(x=data["dilation layer"], y=means, yerr=stds, fmt = 'o',color = 'red', 
            ecolor = '#ffcccb', elinewidth = 5, capsize=10)
plt.ylabel("Pearsonr counts")
plt.xlabel("Dilation Layers")
plt.ylim((0,1))
plt.savefig("figure_2_supp/pearsonr_dilation.pdf",format='pdf',dpi=300, transparent=True)
```
