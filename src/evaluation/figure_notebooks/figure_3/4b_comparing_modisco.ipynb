```python
import pandas as pd
```

```python
dnase_name_counts=pd.read_csv("configs/data/counts.tomtom_dnase.tsv",sep="\t")
atac_name_counts=pd.read_csv("configs/data/counts.tomtom.tsv",sep="\t")
dnase_name_counts=pd.read_csv("configs/data/profile.tomtom_dnase.tsv",sep="\t")
atac_name_counts=pd.read_csv("configs/data/profile.tomtom.tsv",sep="\t")
```

```python
dnase_rename = {"CTCF.1":"CTCF","CTCF.2":"CTCF","FOSL1+JUN\n":"FOSL1+JUN","ATF1":"AP1","ATF1.1":"AP1","BACH.1":"AP1","BACH":"AP1","ETV4":"ELK","ZN143":"ZNF76","JUND":"AP1","CEBPG":"AP1"}
atac_rename = {"BACH.1":"AP1","CTCF.2":"CTCF","FOSL1+JUN\n":"FOSL1+JUN","ATF1":"AP1","ATF1.1":"AP1","BACH":"AP1","ETV4":"ELK","ZN143":"ZNF76","JUND":"AP1","CEBPG":"AP1"}
```

```python
dnase_name_counts=dnase_name_counts[~dnase_name_counts['Pattern'].str.contains("metacluster_1",case=False)] 
atac_name_counts=atac_name_counts[~atac_name_counts['Pattern'].str.contains("metacluster_1",case=False)] 
#dnase_name_profile=dnase_name_profile[~dnase_name_profile['Pattern'].str.contains("metacluster_1",case=False)] 
#atac_name_profile=atac_name_profile[~atac_name_profile['Pattern'].str.contains("metacluster_1",case=False)] 
```

```python
atac_name_counts.head()
```

```python
dnase_name_counts.head()
```

```python
dnase_name_counts["Label"].replace(dnase_rename, inplace=True)
atac_name_counts["Label"].replace(atac_rename, inplace=True)
```

```python
temp = dnase_name_counts.groupby(["Label"]).sum()
dnase_name_counts = temp.loc[:,temp.columns.isin(["Label","Num_Seqlets"])]
dnase_name_counts.loc[:,"Num_Seqlets"] = dnase_name_counts["Num_Seqlets"]/sum(dnase_name_counts["Num_Seqlets"])
```

```python
temp = atac_name_counts.groupby(["Label"]).sum()
atac_name_counts = temp.loc[:,temp.columns.isin(["Label","Num_Seqlets"])]
atac_name_counts.loc[:,"Num_Seqlets"] = atac_name_counts["Num_Seqlets"]/sum(atac_name_counts["Num_Seqlets"])
```

```python
dnase_name_counts.sort_values(by=["Num_Seqlets"],ascending=False)
#dnase_name_counts = dnase_name_counts[dnase_name_counts.loc[:,"Num_Seqlets"]>0.005].sort_values(by=["Num_Seqlets"],ascending=False)

#atac_name_counts[atac_name_counts.loc[:,"Num_Seqlets"]>0.01]
```

```python
atac_name_counts = atac_name_counts.sort_values(by=["Num_Seqlets"],ascending=False)
```

```python
merged = dnase_name_counts.merge(atac_name_counts, on="Label", how="outer").dropna()
merged.columns = ["DNASE", "ATAC"]
```

```python
merged_new=pd.DataFrame()
merged_new["Label"]=merged.index.tolist()+merged.index.tolist()
merged_new["Source"]=["ATAC"]*merged["ATAC"].shape[0]+["DNASE"]*merged["DNASE"].shape[0]
merged_new["Value"]=merged["ATAC"].tolist()+merged["DNASE"].tolist()
```

```python
merged_new.to_csv("configs/data/atac_versus_dnase_profile_compare.tsv",sep=",",header=False,index=False)
```
