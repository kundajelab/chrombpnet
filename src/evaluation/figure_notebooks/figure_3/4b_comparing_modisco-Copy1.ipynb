```python
import pandas as pd
```

```python
# dnase_name_counts=pd.read_csv("configs/data/counts.tomtom_dnase.tsv",sep="\t")
# atac_name_counts=pd.read_csv("configs/data/counts.tomtom.tsv",sep="\t")
dnase_name_counts=pd.read_html("/oak/stanford/groups/akundaje/projects/chromatin-atlas-2022/chrombpnet/folds/DNASE/K562/merge_folds_new_may_05_24/profile/motifs.html")
atac_name_counts=pd.read_html("/oak/stanford/groups/akundaje/projects/chromatin-atlas-2022/chrombpnet/folds/ATAC/K562/merge_folds_new_may_05_24/profile/motifs.html")
```

```python
dnase_name_counts[0]
```

```python

dnase_name_counts=dnase_name_counts[0][~dnase_name_counts[0]['pattern'].str.contains("neg",case=False)] 
atac_name_counts=atac_name_counts[0][~atac_name_counts[0]['pattern'].str.contains("neg",case=False)] 
```

```python
dnase_name_counts["pattern_short"] = "0_"+dnase_name_counts['pattern'].str.split("_").str[2]
atac_name_counts["pattern_short"] = "0_"+atac_name_counts['pattern'].str.split("_").str[2]
```

```python
atac_name_counts = atac_name_counts[["pattern_short", "num_seqlets"]]
dnase_name_counts = dnase_name_counts[["pattern_short", "num_seqlets"]]
```

```python
atac_labels = pd.read_csv("configs/nov_14_2024/atac_profile_annotations.tsv", sep=",", names=["pttns","Label"])
dnase_labels = pd.read_csv("configs/nov_14_2024/dnase_profile_annotations.tsv", sep=",", names=["pttns","Label"])
```

```python
atac_name_counts = atac_name_counts.merge(atac_labels, right_on=["pttns"], left_on=["pattern_short"], how="left")[["pattern_short", "num_seqlets", "Label"]]
dnase_name_counts = dnase_name_counts.merge(dnase_labels, right_on=["pttns"], left_on=["pattern_short"], how="left")[["pattern_short", "num_seqlets", "Label"]]

atac_name_counts
```

```python
# dnase_name_counts["Label"].replace(dnase_rename, inplace=True)
# atac_name_counts["Label"].replace(atac_rename, inplace=True)
```

```python
temp = dnase_name_counts.groupby(["Label"]).sum()
dnase_name_counts = temp.loc[:,temp.columns.isin(["Label","num_seqlets"])]
dnase_name_counts.loc[:,"num_seqlets"] = dnase_name_counts["num_seqlets"]/sum(dnase_name_counts["num_seqlets"])
```

```python
temp = atac_name_counts.groupby(["Label"]).sum()
atac_name_counts = temp.loc[:,temp.columns.isin(["Label","num_seqlets"])]
atac_name_counts.loc[:,"num_seqlets"] = atac_name_counts["num_seqlets"]/sum(atac_name_counts["num_seqlets"])
```

```python
dnase_name_counts.sort_values(by=["num_seqlets"],ascending=False)
#dnase_name_counts = dnase_name_counts[dnase_name_counts.loc[:,"Num_Seqlets"]>0.005].sort_values(by=["Num_Seqlets"],ascending=False)

#atac_name_counts[atac_name_counts.loc[:,"Num_Seqlets"]>0.01]
```

```python
atac_name_counts = atac_name_counts.sort_values(by=["num_seqlets"],ascending=False)
```

```python
atac_name_counts
```

```python
merged = dnase_name_counts.merge(atac_name_counts, on="Label", how="outer").dropna()
merged.columns = ["DNASE", "ATAC"]
```

```python
merged_new=pd.DataFrame()
merged_new["Label"]=merged.index.tolist()+merged.index.tolist()
merged_new["Source"]=["ATAC"]*merged["ATAC"].shape[0]+["DNASE"]*merged["DNASE"].shape[0]
merged_new["Value"]=merged["ATAC"].tolist()+merged["DNASE"].tolist()
```

```python
merged_new.to_csv("configs/data/atac_versus_dnase_profile_compare_nov_2024.tsv",sep=",",header=False,index=False)
merged_new
```
