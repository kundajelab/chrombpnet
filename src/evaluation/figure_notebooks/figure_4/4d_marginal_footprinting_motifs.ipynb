```python
import deepdish
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, inset_axes 
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
import matplotlib

matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42 
```

```python
def normalize_marginal_footprints(input_f):
    for key in input_f:
        if key=="control":
            continue
        input_f[key][0] = input_f[key][0]/sum(input_f[key][0]) 
        #input_f[key][0] = (input_f[key][0]/sum(input_f[key][0])) * (np.log((input_f[key][1]+1)/(input_f["control"][1]+1)))
        #input_f[key][0] = input_f[key][0] * (np.log(input_f[key][1]/input_f["control"][1]))

    return input_f
```

```python
corrected_500M = normalize_marginal_footprints(deepdish.io.load(data_500M_corrected))
uncorrected_500M = normalize_marginal_footprints(deepdish.io.load(data_500M_uncorrected))

corrected_250M = normalize_marginal_footprints(deepdish.io.load(data_250M_corrected))
uncorrected_250M = normalize_marginal_footprints(deepdish.io.load(data_250M_uncorrected))

corrected_100M = normalize_marginal_footprints(deepdish.io.load(data_100M_corrected))
uncorrected_100M = normalize_marginal_footprints(deepdish.io.load(data_100M_uncorrected))

corrected_50M = normalize_marginal_footprints(deepdish.io.load(data_50M_corrected))
uncorrected_50M = normalize_marginal_footprints(deepdish.io.load(data_50M_uncorrected))

corrected_25M = normalize_marginal_footprints(deepdish.io.load(data_25M_corrected))
uncorrected_25M = normalize_marginal_footprints(deepdish.io.load(data_25M_uncorrected))

corrected_5M = normalize_marginal_footprints(deepdish.io.load(data_5M_corrected))
uncorrected_5M = normalize_marginal_footprints(deepdish.io.load(data_5M_uncorrected))



```

```python
import numpy as np
def make_inset(input_array, main_axes, color, ymax, ymin, yvis, start=500-50+5, end=500+50+5):
    x1 = 0
    x2 = 100

    input_array = input_array[start:end]
    input_array = input_array*10000

    # select y-range for zoomed region
    #y1 = np.floor(min(input_array))
    #y2 = np.ceil(max(input_array))
    
    y1=np.floor(ymin*10000)
    y2=np.ceil(ymax*10000)
    
    #print(max(input_array))
    

    # Make the zoom-in plot:
    #  axins = zoomed_inset_axes(axs[i-1,3], 2, loc=1) # zoom = 2
    axins = inset_axes(main_axes, loc=1, width="30%", height=0.6) # zoom = 2

    axins.plot(input_array, c=color, alpha=0.8)
    axins.set_xlim(x1, x2)
    axins.set_ylim(y1, y2)
    #axins.set_xticks([80, 120])

    axins.tick_params(axis='both', which='minor', labelsize=8)

    plt.locator_params(axis='y', nbins=2)
    axins.set_yticks([y1, y2])

    plt.xticks(visible=False)
    plt.yticks(visible=yvis)

    #(main_axes, axins, loc1=2, loc2=4, fc="none", ec="0.5", linewidth=0.1, linestyle="--")
    plt.draw()
    
    
```

```python
def plot_tracks(uncorrected, corrected, ax=None, ylim=0.01, start=500-100+5, end=500+100+5 ):
    width = end - start
    ax.plot(range(width), uncorrected[start:end]*1000, 'r', linewidth=1, alpha=0.3, label="uncorrected")
    ax.plot(range(width), corrected[start:end]*1000, 'k', linewidth=1, label="corrected")
    ax.spines[['top', 'right']].set_visible(False)

    #ax.set_ylim(0,ylim) 
```
