```python
atac_model=pd.read_csv("/mnt/lab_data2/anusri/enformer/eu_caqtls/chrombpnet_preds_asb/atac/ATAC/summary.mean.variant_scores_new_2.tsv", sep='\t', header=0)
atac_model_50m=pd.read_csv("/mnt/lab_data2/anusri/enformer/eu_caqtls/chrombpnet_preds_asb/atac_50M/ATAC_50M/summary.mean.variant_scores_new_2.tsv", sep='\t', header=0)
dnase_model=pd.read_csv("/mnt/lab_data2/anusri/enformer/eu_caqtls/chrombpnet_preds_asb/dnase_predss/DNASE/summary.mean.variant_scores_new_2.tsv", sep='\t', header=0)
```

```python
vals = ['GM12878.ENCSR637XSC.mean.logfc', 'GM12878.ENCSR637XSC.mean.abs_logfc', 'GM12878.ENCSR637XSC.mean.jsd',
       'GM12878.ENCSR637XSC.mean.abs_logfc_x_jsd',
       'GM12878.ENCSR637XSC.mean.abs_logfc_x_jsd_x_max_percentile',
       'GM12878.ENCSR637XSC.mean.max_percentile',
       'GM12878.ENCSR637XSC.mean.abs_logfc.pval',
       'GM12878.ENCSR637XSC.mean.jsd.pval',
       'GM12878.ENCSR637XSC.mean.abs_logfc_x_jsd.pval',
       'GM12878.ENCSR637XSC.mean.abs_logfc_x_jsd_x_max_percentile.pval']

for val in vals:
    print(val)
    if "pval" not in val:
        asb_sig[val] = atac_model[val.replace("GM12878.ENCSR637XSC.mean.", "")+".mean"]
    else:
        asb_sig[val] = atac_model[val.replace("GM12878.ENCSR637XSC.mean.", "").replace(".pval",".mean.pval")]
```

```python
for val in vals:
    print(val)
    if "pval" not in val:
        asb_sig[val.replace("ENCSR637XSC", "ENCSR000EMT")] = dnase_model[val.replace("GM12878.ENCSR637XSC.mean.", "")+".mean"]
    else:
        asb_sig[val.replace("ENCSR637XSC", "ENCSR000EMT")] = dnase_model[val.replace("GM12878.ENCSR637XSC.mean.", "").replace(".pval",".mean.pval")]
```

```python
for val in vals:
    print(val)
    if "pval" not in val:
        asb_sig[val.replace("ENCSR637XSC", "ENCSR637XSC.sub.50M")] = atac_model_50m[val.replace("GM12878.ENCSR637XSC.mean.", "")+".mean"]
    else:
        asb_sig[val.replace("ENCSR637XSC", "ENCSR637XSC.sub.50M")] = atac_model_50m[val.replace("GM12878.ENCSR637XSC.mean.", "").replace(".pval",".mean.pval")]
```

```python
anscs = asb_sig.columns[5:70].tolist()+asb_sig.columns[72:].tolist()
anscs = [ans.split(".")[0] for ans in anscs]
anscs = set(anscs)
print(anscs)

sub_list = []
for ans in anscs:
    if ans == "GM12878":
        for title in [".ENCSR637XSC", ".ENCSR000EMT", ".ENCSR637XSC.sub.50M", ".ENCSR095QNB"]:
            sub_list.append(ans+title+".mean.logfc")
            asb_sig[ans+title+".mean.sign.jsd"] = np.sign(asb_sig[ans+title+".mean.logfc"]) * asb_sig[ans+title+".mean.jsd"]
            sub_list.append(ans+title+".mean.sign.jsd")
            sub_list.append(ans+title+".mean.abs_logfc_x_jsd_x_max_percentile")
            sub_list.append(ans+title+".mean.abs_logfc_x_jsd_x_max_percentile.pval")
    else:
        sub_list.append(ans+".mean.logfc")
        asb_sig[ans+".mean.sign.jsd"] = np.sign(asb_sig[ans+".mean.logfc"]) * asb_sig[ans+".mean.jsd"]
        sub_list.append(ans+".mean.sign.jsd")        

    
```

```python
asb_sig = asb_sig[asb_sig.columns[0:5].tolist()+asb_sig.columns[70:72].tolist()+sub_list]
```

```python
asb_sig.head()
```

```python
asb_sig["pos0"] = asb_sig["pos"] - 1
selected_beds = asb_sig[["chr", "pos0", "pos"]+asb_sig.columns[2:-1].tolist()]
selected_beds.head()
```

```python
import pybedtools
beds_var = pybedtools.BedTool.from_dataframe(selected_beds)
peak_bed = pybedtools.BedTool('af_caqtls/merged.peaks.af.bed')
sig_vars_in_bed = beds_var.intersect(peak_bed, wa=True)
```

```python
sig_vars_in_bed = sig_vars_in_bed.to_dataframe(names=["chr", "pos0", "pos"]+asb_sig.columns[2:-1].tolist()).drop_duplicates()
```

```python
asb_sig["IsUsed"] = asb_sig["rsid"].isin(sig_vars_in_bed["rsid"])
```

```python
# the ground truths are flipped when rsid does not match allele2 and allele
dataf = asb_sig
index=(dataf["allele1"] != (dataf["rsid"].str.split('_').str[2]))
index=np.array(index.values)
for ans in anscs:
    if ans=="eur":
        continue
    if ans=="afr":
        continue       
    print(ans)
    if ans == "GM12878":
        for title in [".ENCSR637XSC", ".ENCSR000EMT", ".ENCSR637XSC.sub.50M", ".ENCSR095QNB" ]:
            print(title)
            dataf.loc[index,ans+title+".mean.sign.jsd"] = -1 * dataf.loc[index,ans+title+".mean.sign.jsd"]
            dataf.loc[index,ans+title+".mean.logfc"] = -1 * dataf.loc[index,ans+title+".mean.logfc"]
    else:
        title=""
        dataf.loc[index,ans+title+".mean.sign.jsd"] = -1 * dataf.loc[index,ans+title+".mean.sign.jsd"]
        dataf.loc[index,ans+title+".mean.logfc"] = -1 * dataf.loc[index,ans+title+".mean.logfc"]
        
```
