```python
import pysam 
import pandas as pd
ref=pysam.FastaFile('/mnt/data/male.hg19.fa')
chromsizes=pd.read_csv('/mnt/data/annotations/by_release/hg19/hg19.chrom.sizes',header=None,nrows=24,sep='\t')
chromsizes
```

```python
seq=''
for index,row in chromsizes.iterrows(): 
    cur_chrom=row[0]
    cur_size=row[1] 
    seq=seq+ref.fetch(cur_chrom,0,cur_size).lower()
```

```python
len(seq)
```

```python
a_count=seq.count('a')
t_count=seq.count('t')
g_count=seq.count('g')
c_count=seq.count('c')
```

```python
total_bases=a_count+c_count+g_count+t_count
```

```python
a_frac=a_count/len(seq) 
c_frac=c_count/len(seq)
g_frac=g_count/len(seq) 
t_frac=t_count/len(seq) 
```
