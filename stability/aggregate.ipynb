```r
library(reshape2)
library(ggplot2)
```

```r
all_spearman=read.table("spearman.tsv",header=TRUE,sep='\t')
all_spearman=all_spearman[all_spearman$fold1!=all_spearman$fold2,]
all_spearman$fold1=NULL
all_spearman$fold2=NULL 
head(all_spearman)
```

```r
colnames(all_spearman)=c("ProfileSHAP","CountSHAP","BinnedRegressionSHAP","BinnedClassificationSHAP","GKMExplain")
```

```r
all_spearman=melt(all_spearman)
head(all_spearman)
```

```r
options(repr.plot.width=10, repr.plot.height=4)
ggplot(all_spearman, aes(value,group=variable,color=variable)) + 
    stat_ecdf(geom = "step")+
    scale_color_manual(values=c('#e41a1c','#377eb8','#4daf4a','#984ea3','#ff7f00'),name="Interpretation Scores")+
    xlab("Spearman Across Folds")+
    ylab("CDF")+
    theme_bw(20)
```
