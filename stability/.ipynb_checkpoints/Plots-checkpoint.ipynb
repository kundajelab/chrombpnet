```r
library(ggplot2)
library(ggridges)
source("~/helpers.R")
library(reshape2)
```

```r
data=read.table("cosine_sim.tsv",header=TRUE,sep='\t')
data$same_fold=FALSE
data$same_fold[data$fold1==data$fold2]=TRUE
print(nrow(data))
data=data[data$same_fold==FALSE,]
print(nrow(data))
names(data)=c("BPNetProfile","BPNetCounts","BinaryClass","BinaryReg","GKMExplain","fold1","fold2")
data=melt(data,id=c("fold1","fold2"))
head(data)
```

```r
options(repr.plot.width=20, repr.plot.height=20)

p1=ggplot(data,
       aes(y=variable,
           x=value))+
  geom_density_ridges(
    jittered_points=TRUE,
    scale=0.95,
    rel_min_height=0.01,
    point_shape = "|", point_size = 3, size = 0.25,
    position = position_points_jitter(height = 0)
  ) +
  xlim(0.5,1.2)+
  scale_y_discrete(expand = c(0, 0), name="deepSHAP model") +
  coord_cartesian(clip = "off") +
  ggtitle("Cosine Similarity, K562 DNASE") +
  theme_bw(20)
```
