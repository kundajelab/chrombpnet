```r
library(reshape2)
library(ggplot2)
```

```r
all_spearman=read.table("spearman.tsv",header=TRUE,sep='\t')
all_spearman=all_spearman[all_spearman$fold1!=all_spearman$fold2,]
all_spearman$fold1=NULL
all_spearman$fold2=NULL 
all_spearman=melt(all_spearman)
head(all_spearman)
```

```r
ggplot(all_spearman, aes(value,group=variable,color=variable)) + stat_ecdf(geom = "step")
```

```r
all_cosine_sim=read.table("cosine_sim.tsv",header=TRUE,sep='\t')
all_cosine_sim=all_cosine_sim[all_cosine_sim$fold1!=all_cosine_sim$fold2,]

head(all_cosine_sim)
```
