## JSD quantiles


```python

%load_ext autoreload
%autoreload 2
import random
import pandas as pd 
from scipy.special import softmax 
import h5py
import pyBigWig

import sys 
sys.path.append("..")
from helpers import *

## for plotting 
import matplotlib 
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"]=10,10
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 10}

matplotlib.rc('font', **font)
```

```python
jsd_bpnet=pd.read_csv("../bpnet.bias.preds.k562.0.scores.0.jsd.txt",header=0,sep='\t',index_col=[0,1])
jsd_bpnet.head()
```

```python
# quantile cutoffs 
quantile_0=jsd_bpnet.JSD.quantile(0.0)
print(quantile_0)
quantile_20=jsd_bpnet.JSD.quantile(.20)
print(quantile_20)
quantile_40=jsd_bpnet.JSD.quantile(.40)
print(quantile_40)
quantile_60=jsd_bpnet.JSD.quantile(.60)
print(quantile_60)
quantile_80=jsd_bpnet.JSD.quantile(.80)
print(quantile_80)
quantile_100=jsd_bpnet.JSD.quantile(1)
print(quantile_100)
```

```python
#get JSD values by quantile/region 
jsd_bpnet_20=jsd_bpnet[jsd_bpnet.JSD.between(quantile_0,quantile_20) ]
jsd_bpnet_40=jsd_bpnet[jsd_bpnet.JSD.between(quantile_20,quantile_40) ]
jsd_bpnet_60=jsd_bpnet[jsd_bpnet.JSD.between(quantile_40,quantile_60) ]
jsd_bpnet_80=jsd_bpnet[jsd_bpnet.JSD.between(quantile_60,quantile_80) ]
jsd_bpnet_100=jsd_bpnet[jsd_bpnet.JSD.between(quantile_80,quantile_100) ]
```

```python
#randomly select 10 examples per quantile 
random.seed(1234)
jsd_bpnet_20_sample=jsd_bpnet_20.sample(n=10)
jsd_bpnet_40_sample=jsd_bpnet_40.sample(n=10)
jsd_bpnet_60_sample=jsd_bpnet_60.sample(n=10)
jsd_bpnet_80_sample=jsd_bpnet_80.sample(n=10)
jsd_bpnet_100_sample=jsd_bpnet_100.sample(n=10)
```

## load the model labels & predictions 

```python
#load the predictions from bpnet,tobias,and 6mer bias models 
preds_bpnet=h5py.File("../idr_preds/bpnet.bias.preds.k562.0.predictions",'r')
preds_tobias=h5py.File("../idr_preds/tobias.bias.preds.k562.0.predictions",'r')
preds_6mer=h5py.File("../idr_preds/6mer.bias.preds.k562.0.predictions",'r')
```

```python
coords=preds_bpnet['coords'][:]
coords=[tuple([i.decode('utf8')  for i in j[0:2]]) for j in coords]
coord_to_index={}  
for i in range(len(coords)): 
    cur_chrom=coords[i][0] 
    cur_pos=int(coords[i][1])
    coord_to_index[tuple([cur_chrom,cur_pos])]=i
```

```python
coord_to_index
```

```python
coords_q20=[coord_to_index[i] for i in jsd_bpnet_20_sample.index]
coords_q40=[coord_to_index[i] for i in jsd_bpnet_40_sample.index]
coords_q60=[coord_to_index[i] for i in jsd_bpnet_60_sample.index]
coords_q80=[coord_to_index[i] for i in jsd_bpnet_80_sample.index]
coords_q100=[coord_to_index[i] for i in jsd_bpnet_100_sample.index]
```

```python
#coordinates in increasing order
coords_q20.sort()
coords_q40.sort()
coords_q60.sort()
coords_q80.sort()
coords_q100.sort()
```

## q20 

```python
q20_bpnet_toplot=get_model_outputs_to_plot(preds_bpnet,coords_q20)
q20_tobias_toplot=get_model_outputs_to_plot(preds_tobias,coords_q20)
q20_6mer_toplot=get_model_outputs_to_plot(preds_6mer,coords_q20)
```

```python
plt.rcParams["figure.figsize"]=35,16
for i in range(1,2): 
    (chrom,summit)=jsd_bpnet_20_sample.index[i]
    print(chrom)
    print(summit)
    fig, axes = plt.subplots(13, 1)
    #labels 
    axes[0].plot(q20_bpnet_toplot['labels_counts'][i],label='Labels BPNET',color='r')
    axes[0].plot(q20_tobias_toplot['labels_counts'][i],label='Labels TOBIAS',color='g')
    axes[0].plot(q20_6mer_toplot['labels_counts'][i],label='Labels Vierstra',color='b')

    #predict counts 
    axes[1].plot(q20_bpnet_toplot['predictions_counts'][i],label='Pred Counts BPNET',color='r')
    axes[1].plot(q20_tobias_toplot['predictions_counts'][i],label='Pred Counts TOBIAS',color='g')
    axes[1].plot(q20_6mer_toplot['predictions_counts'][i],label='Pred Counts Vierstra',color='b')

    #observed - predicted (logit space)
    axes[2].plot(q20_bpnet_toplot['delta_logits'][i],label='Logit Lab - Logit Pred BPNET',color='r')
    axes[2].plot(q20_tobias_toplot['delta_logits'][i],label='Logit Lab - Logit Pred TOBIAS',color='g')
    axes[2].plot(q20_6mer_toplot['delta_logits'][i],label='Logit Lab - Logit Pred Vierstra',color='b')
        
    #Prob (logit observed - logit predicted)
    axes[3].plot(q20_bpnet_toplot['delta_prob'][i],label='Prob(Logit Lab - Logit Pred) BPNET',color='r')
    axes[3].plot(q20_tobias_toplot['delta_prob'][i],label='Prob(Logit Lab - Logit Pred) TOBIAS',color='g')
    axes[3].plot(q20_6mer_toplot['delta_prob'][i],label='Prob(Logit Lab - Logit Pred) Vierstra',color='b')
        
    #Tobias footprints
    axes[4].plot(np.nan_to_num(tobias_footprints.values(chrom,summit-500,summit+500,numpy=True)),label="TOBIAS Footprints",color='k')
    
    #Tobias uncorrected 
    axes[5].plot(np.nan_to_num(tobias_uncorrected.values(chrom,summit-500,summit+500,numpy=True)),label="TOBIAS Uncorrected",color='k')

    #Tobias bias 
    axes[6].plot(np.nan_to_num(tobias_bias.values(chrom,summit-500,summit+500,numpy=True)),label="TOBIAS Bias",color='k')

    #Tobias expected 
    axes[7].plot(np.nan_to_num(tobias_expected.values(chrom,summit-500,summit+500,numpy=True)),label="TOBIAS Expected",color='k')
    
    #Tobias corrected
    axes[8].plot(np.nan_to_num(tobias_corrected.values(chrom,summit-500,summit+500,numpy=True)),label="TOBIAS Corrected",color='k')
    
    #Vierstra Kmer 
    axes[9].plot(np.nan_to_num(vierstra_kmers.values(chrom,summit-500,summit+500,numpy=True)),label="Vierstra Kmers",color='k')
    
    #Vierstra Kmer Smoothed
    axes[10].plot(np.nan_to_num(vierstra_kmers_smoothed.values(chrom,summit-500,summit+500,numpy=True)),label="Vierstra Kmers Smoothed",color='k')
        
    #Vierstra Footprint 1 
    axes[11].plot(np.nan_to_num(vierstra_fpr1.values(chrom,summit-500,summit+500,numpy=True)),label="Vierstra Footprints 1",color='k')
    
    #Vierstra Footprint 2 
    axes[12].plot(np.nan_to_num(vierstra_fpr2.values(chrom,summit-500,summit+500,numpy=True)),label="Vierstra Footprints 2",color='k')

      
    axes[0].set_title('Q20 K562 DNASE bias models,'+str(chrom)+":"+str(summit))
    axes[0].legend(loc=1)
    axes[1].legend(loc=1) 
    axes[2].legend(loc=1)
    axes[3].legend(loc=1)
    axes[4].legend(loc=1)
    axes[5].legend(loc=1)
    axes[6].legend(loc=1)
    axes[7].legend(loc=1)
    axes[8].legend(loc=1)
    axes[9].legend(loc=1)
    axes[10].legend(loc=1)
    axes[11].legend(loc=1)
    axes[12].legend(loc=1)
    
```
