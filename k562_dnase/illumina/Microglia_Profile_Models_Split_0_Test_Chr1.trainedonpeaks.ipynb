## summits

```python
import pandas as pd 
```

```python
labels_peakonly=pd.read_hdf("0.task0.predictions.summits.microglia.0.chr1.trainedonpeaks.labels",header=None,sep='\t')
labels=pd.read_hdf("0.task0.predictions.summits.microglia.0.chr1.labels",header=None,sep='\t')
```

```python
predictions_peakonly=pd.read_hdf("0.task0.predictions.summits.microglia.0.chr1.trainedonpeaks.predictions",header=None,sep='\t')
predictions=pd.read_hdf("0.task0.predictions.summits.microglia.0.chr1.predictions",header=None,sep='\t')
```

```python
print(labels.shape)
print(labels_peakonly.shape)
```

```python
print(predictions.shape)
print(predictions_peakonly.shape)
```

```python
indices=labels.index.tolist()
```

```python
import matplotlib 
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"]=10,8
```

```python
for i in range(20):
    cur_index=indices[i]
    plt.figure()
    plt.plot(predictions.loc[cur_index],label='predictions')
    plt.plot(labels.loc[cur_index],linestyle='dashed',label='labels')
    plt.plot(predictions_peakonly.loc[cur_index],label='predictions.trained.on.peaks')
    plt.plot(labels_peakonly.loc[cur_index],linestyle='dashed',label='labels.trained.on.peaks')
    plt.title(str(cur_index))
    plt.legend()
```
