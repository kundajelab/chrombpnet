## summits

```python
import pandas as pd 
```

```python
labels=pd.read_hdf("predictions.k562.0.chr1.labels.0",header=None,sep='\t')
```

```python
predictions=pd.read_hdf("predictions.k562.0.chr1.predictions.0",header=None,sep='\t')
```

```python
print(labels.shape)
```

```python
print(predictions.shape)
```

```python
indices=labels.index.tolist()
```

```python
import matplotlib 
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"]=10,8
```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)
for i in range(20):
    cur_index=indices[i]
    plt.figure()
    plt.plot(predictions.loc[cur_index],label='Predictions')
    plt.plot(labels.loc[cur_index],linestyle='dashed',label='Labels')
    plt.title('K562 DNAse '+str(cur_index))
    plt.legend(loc='best')
```

## genomewide

```python
labels=pd.read_hdf("predictions.k562.0.chr1.genomewide.labels.0",header=None,sep='\t')
predictions=pd.read_hdf("predictions.k562.0.chr1.genomewide.predictions.0",header=None,sep='\t')
indices=labels.index.tolist()
```

```python
for i in range(20):
    cur_index=indices[i]
    plt.figure()
    plt.plot(predictions.loc[cur_index],label='predictions')
    plt.plot(labels.loc[cur_index],linestyle='dashed',label='labels')
    plt.title(str(cur_index))
    plt.legend()
```
